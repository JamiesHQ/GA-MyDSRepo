#### 10. Display a histogram of ages of all killings


```python
import matplotlib.pyplot as plt
%matplotlib inline

#using http://matplotlib.org/examples/statistics/histogram_demo_histtypes.html
killings['age'].hist()
plt.show()
```

```python
#with seaborn
import seaborn as sns
sns.set(color_codes=True)
plt.hist(killings['age'])
plt.show()
```

```python
#Official Answer
fig = plt.figure(figsize=(15,10))
ax  = fig.gca()

killings.groupby('age')['Unnamed: 0'].count().reset_index().sort_values('age', ascending=True).plot(kind='bar', x='age', ax=ax)
ax.legend_.remove()
```
